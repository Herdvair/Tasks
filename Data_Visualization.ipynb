Saya mengidentifikasi missing value di kolom total charge dengan indikasi data yang tidak sesuai dalam kolom tersebut. Lalu saya menghitung persentase missing valuesnya  dan mencapai 0.1561%. Meskipun tergolong kecil, namun tetap dikatakan missing values karena persentasenya bukan 0%.

```python
mode_tc = df_telco[~df_telco['Total Charges'].isin([' '])]['Total Charges'].mode()[0] #menggunakan modus karena tipe data object
mode_tc
```

```python
df_telco['Total Charges'] = df_telco['Total Charges'].apply(lambda x: mode_tc if x in [' '] else x)
```

```python
from scipy import stats
from scipy.stats import skew,shapiro
```

```python
distribusi_col = ['Tenure Months', 'Monthly Charges']

for i in distribusi_col:
    distribusi = df_telco[i]
    nilai_skew = skew(distribusi)
    stat, p = shapiro(distribusi)

    print(f"kolom : {i}")
    print(f"memiliki skewness {nilai_skew}")
    print(f"Hasil uji saphiro, Nilai stat : {stat :.4f}, nilai p-value : {p :.4f}")

    if p > 0.05:
        print("distribusi data normal")
    else:
        print("distribusi data tidak normal")

    sns.histplot(distribusi, kde=True)

    plt.title('Distribusi')
    plt.xlabel(i)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
correlation_matrix = np.corrcoef(df_telco['Tenure Months'], df_telco['Monthly Charges'])
correlation_matrix
```

Korelasi antara "Tenure Months" dan "Monthly Charges" menunjukkan korelasi posiitf sempurna, namun kekuatan korelasinya lemah.

```python
plt.scatter(df_telco['Tenure Months'], df_telco['Monthly Charges'])
plt.show()
```

**Interpretasi :**

Karena korelasi antara Tenure Months dan Monthly Charges bernilai positif lemah (sekitar 0.247), dapat dikatakan bahwa ada pengaruh lamanya durasi langganan bulanan pelanggan terhadap pengeluaran (biaya) yang dikeluarkan oleh pelanggan tiap bulan, namun bukan hanya faktor dari "Tenure Months" bisa jadi ada faktor lain yang memengaruhi pengeluaran pelanggan tiap bulan,

```python
# melakukan normalisasi data unntuk data kategopri Streaming Movies
df_telco['Streaming Movies'] = df_telco['Streaming Movies'].replace({'No internet service' : 'No'})
```

```python
top_cities =df_telco['City'].value_counts().nlargest(5).index
top_cities
```

```python
df_topcity = df_telco[df_telco['City'].isin(top_cities)]
```

```python
cross_tab = pd.crosstab(df_topcity['City'], df_topcity['Streaming Movies'])
```

```python
categories = cross_tab.index.tolist()      # City
subcategories = cross_tab.columns.tolist() # Yes/No
values = cross_tab.values                  # matrix nilai

fig, ax = plt.subplots()
bottom_values = np.zeros(len(categories))

for i, subcategory in enumerate(subcategories):
    ax.bar(categories, values[:, i], bottom=bottom_values, label=subcategory)
    bottom_values += values[:, i]

ax.set_ylabel('Jumlah')
ax.set_title('Perbandingan Streaming Movies berdasarkan Kota')
ax.legend(title='Streaming Movies')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
