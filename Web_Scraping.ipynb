## Data Scraping Menggunakan BeautifulSoup

```python
# API URL
url_hotel = 'https://www.oyorooms.com/search?location=Makassar%2C%20South%20Sulawesi%2C%20Indonesia&city=Makassar&searchType=city&coupon=&checkin=07%2F07%2F2025&checkout=14%2F07%2F2025&roomConfig%5B%5D=2&showSearchElements=false&guests=2&rooms=1&countryName=Indonesia&country=indonesia&filters%5Bcity_id%5D=700'
option = webdriver.ChromeOptions()
option.add_argument("--start-maximized")

# uncomment untuk running code dibawah ini jika run di google collab
option.add_argument("--headless")
option.add_argument("--disable-gpu")
option.add_argument("--no-sandbox")
option.add_argument("--remote-debugging-port=9230")
# comment 4 code diatas (option.add) jika running di local / jupyter notebook dsb
driver_hotel = webdriver.Chrome(options=option)
driver_hotel.get(url_hotel)
# Membuat objek BeautifulSoup dari konten HTML
soup = BeautifulSoup(driver_hotel.page_source, "html.parser")
```

```python
for tag in soup.find_all(True):
    print(tag.name)
```

```python
soup
```

```python
div_containers_oyo = soup.find_all("div", attrs={'class' :'oyo-cell--12-col oyo-cell--8-col-tablet oyo-cell--4-col-phone'})
print(div_containers_oyo)
```

### One Page

1. Nama Hotel

```python
hotel_oyo = []
for container in div_containers_oyo:
    list_nama_oyo = container.find('h3', attrs={"class":'listingHotelDescription__hotelName d-textEllipsis'})
    if list_nama_oyo:
      nama_oyo = list_nama_oyo.text.strip()
    else:
      nama_oyo = None
    hotel_oyo.append(nama_oyo)

hotel_oyo = pd.DataFrame(hotel_oyo, columns=['Name'])
hotel_oyo
```

2. Harga

```python
harga_hotel_oyo = []
for container in div_containers_oyo:
    list_harga_oyo = container.find('span', attrs={"class":'listingPrice__finalPrice listingPrice__finalPrice--black'})
    if list_harga_oyo:
      harga_oyo = list_harga_oyo.text.strip()
    else:
      harga_oyo = None
    harga_hotel_oyo.append(harga_oyo)

harga_hotel_oyo = pd.DataFrame(harga_hotel_oyo, columns=['Price'])
harga_hotel_oyo
```

3. Lokasi

```python
lokasi_hotel_oyo = []
for container in div_containers_oyo:
    list_lokasi_oyo = container.find('span', attrs={"class":'u-line--clamp-2'})
    if list_lokasi_oyo:
      lokasi_oyo = list_lokasi_oyo.text.strip()
    else:
      lokasi_oyo = None
    lokasi_hotel_oyo.append(lokasi_oyo)

lokasi_hotel_oyo = pd.DataFrame(lokasi_hotel_oyo, columns=['Location'])
lokasi_hotel_oyo
```

4. Wizard Member

```python
member_hotel_oyo = []
for container in div_containers_oyo:
    wizard_member_oyo = container.find('div', attrs={"class":'oyo-cell--12-col listingHotelDescription__labels'})
    if wizard_member_oyo:
      member_oyo = wizard_member_oyo.text.strip()
    else:
      member_oyo = None
    member_hotel_oyo.append(member_oyo)

member_hotel_oyo = pd.DataFrame(member_hotel_oyo, columns=['Membership'])
member_hotel_oyo
```

5. Rating

```python
rating_hotel_oyo = []
for container in div_containers_oyo:
    list_rating_oyo = container.find('div', attrs={"class":'hotelRating__wrapper'})
    if list_rating_oyo:
      rating_oyo = list_rating_oyo.text.strip()
    else:
      rating_oyo = None
    rating_hotel_oyo.append(rating_oyo)

rating_hotel_oyo = pd.DataFrame(rating_hotel_oyo, columns=['Rating'])
rating_hotel_oyo
```

### Data Cleaning

```python
df_oyo = pd.concat([hotel_oyo,harga_hotel_oyo, lokasi_hotel_oyo, member_hotel_oyo, rating_hotel_oyo], axis=1)

# Print the total number of hotels
print('Total Hotels : ', len(df_oyo))

# Display the first few rows of the combined DataFrame
df_oyo
```

```python
df_oyo['Price_in_IDR'] = (
    df_oyo['Price']
    .str.replace(r'[^\d]', '', regex=True)
    .astype(int)
)

df_oyo
```

```python
df_oyo['Total_Rating'] = (
    df_oyo['Rating']
    .str.extract(r'(\d+\.\d+)')
    .astype(float)
)
df_oyo
```

```python
#export to csv
df_oyo.to_csv('df_oyo.csv', index=False)
```

```python
div_containers = soup.find_all("div", attrs={'class' :'css-1dbjc4n r-14lw9ot r-1xfd6ze r-da5iq2 r-zo7nv5 r-1loqt21 r-18u37iz r-1otgn73 r-1i6wzkk r-lrvibr r-13qz1uu'})
print(div_containers)
```

### One Page

1. Nama Hotel

```python
name_hotel = []
for container in div_containers:
    name = container.find('h3', attrs={"class":'css-4rbku5'})
    print(name)
    if name:
      names = name.text.strip()
    else:
      names = None
    name_hotel.append(names)

name_hotel = pd.DataFrame(name_hotel, columns=['Name'])
name_hotel
```

2. Harga

```python
# Extract prices
price_hotel = []
for container in div_containers:
    price = container.find('div', attrs={"class": 'css-901oao r-uh8wd5 r-b88u0q r-1ff274t'})
    if price:
       prices = price.text.strip()
    else:
       prices = None
    price_hotel.append(prices)

# Convert price list to DataFrame
price_hotel = pd.DataFrame(price_hotel, columns=['Price'])
price_hotel
```

3. Lokasi

```python
location_hotel = []

for container in div_containers:
    location = container.find('div', attrs={"class": 'css-1dbjc4n r-1awozwy r-1kbdv8c r-18u37iz r-dnmrzs'})
    if location:
        locations = location.text.strip()
    else:
        locations = None

    location_hotel.append(locations)

# Convert to DataFrame
location_hotel = pd.DataFrame(location_hotel, columns=['location'])
location_hotel
```

4. Jaminan refund

```python
jaminan_refund = []

for container in div_containers:
    refund = container.find('div', attrs={"class": 'css-1dbjc4n r-18u37iz r-1udh08x r-1hvjb8t r-oyd9sg'})
    if refund:
        refunds = refund.text.strip()
    else:
        refunds = None

    # Pindahkan append ke dalam loop, dan simpan text
    jaminan_refund.append(refunds)

# Convert to DataFrame
jaminan_refund = pd.DataFrame(jaminan_refund, columns=['Jaminan_Refund'])
jaminan_refund
```

5. Rating

```python
# Extract ratings
ratings_hotel = []
for container in div_containers:
    rating = container.find('div', attrs={"class": 'css-901oao r-b88u0q r-fdjqy7'})
    if rating:  # cek jika elemen ditemukan
        ratings = rating.text.strip()
    else:
        ratings = None  # atau bisa juga kasih default seperti 'N/A'
    ratings_hotel.append(ratings)

# Convert ratings list to DataFrame
ratings_hotel = pd.DataFrame(ratings_hotel, columns=['ratings'])
ratings_hotel
```

```python
df_traveloka = pd.concat([name_hotel, price_hotel, jaminan_refund,ratings_hotel, location_hotel], axis=1)

# Print the total number of hotels
print('Total Hotels : ', len(df_traveloka))

# Display the first few rows of the combined DataFrame
df_traveloka
```
